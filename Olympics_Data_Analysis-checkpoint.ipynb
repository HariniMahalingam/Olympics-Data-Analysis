```python
# importing pandas for reading the data and creating dataframe
import pandas as pd

# Reading the csv file
df = pd.read_csv("athlete_events.csv")
print(df)
```

```python
# metadata of the file
df.info()
```

```python
# Returns the aggregate values of each columns like count,mean,std,min,max,25% etc.
df.describe()
```

```python
# Removes the row if a column or field in a dataframe is empty(NA)
df.dropna()
```

```python
# Removes the row for the specified column or field in a dataframe is empty(NA)
print(df.dropna(subset=['ID']))
print('\n')
print("******************")
print('\n')
print(df.dropna(subset=['Age']))
```

```python
# Replaces all the blank values in all columns in the data frame with 0(value specified in the fillna)
print(df.fillna(0))
print('\n')
print("******************")
print('\n')
# Replaces the blank values in the column specified with the value given
print(df.fillna({'Age':0,'Height':0}))
```

```python
# merges both the dataframes.
datframe = pd.read_csv("noc_regions.csv")
obj=pd.merge(df,datframe)
```

```python
# To retreive the data based on some conditon. I am using this to remove NaN values and analyse only gold medal data
obj[obj['Medal'] == 'Gold']
```

```python
import seaborn as sns

sns.countplot(x=obj['Medal'])
```

```python
import matplotlib.pyplot as plt
gold_medal = obj[obj['Medal'] == 'Gold']
sns.countplot(x=gold_medal['Age'])
plt.figure(figsize=(3383333382,10211133321))
```

```python
# Giving  conditions and filtering the data in a table
obj['Sport'] [obj['Age'] > 50]
```

```python
import matplotlib.pyplot as plt
greater_age = obj['Sport'] [obj['Age'] > 50]
plt.figure(figsize=(20, 10))
sns.countplot(x=greater_age)
```

```python
# Top 5 countries with highest gold medal
goldmedals=obj[obj['Medal'] == 'Gold']
print(goldmedals.region.value_counts().reset_index(name='Medal').head())

totalGoldMedals = goldmedals.region.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
				height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country')
plt.show()
```
